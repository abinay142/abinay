```python
import pandas as pd
```

```python
data = pd.read_csv("data.csv")
data
```

```python
data['Values'] = data['Values'].str.split(';')
```

```python
V1 = []
V2 = []
V3 = []
V4 = []
V5 = []

for i in data['Values']:
    
    V1.append(i[0])
    V2.append(i[1])
    V3.append(i[2])
    V4.append(i[3])
    V5.append(i[4])        
```

```python
data['V1'], data['V2'], data['V3'], data['V4'], data['V5'] = V1, V2, V3, V4, V5
data = data.drop(['Values'], axis = 1)
data
```

```python
data['V1'] = data['V1'].astype(int)
data['V2'] = data['V2'].astype(int)
data['V3'] = data['V3'].astype(int)
data['V4'] = data['V4'].astype(int)
data['V5'] = data['V5'].astype(int)
```

```python
data.dtypes
```

```python
data_V2 = data.groupby('Country').sum()
data_V3 = data_V2.reset_index()
data_V3
```

```python
data_V3['Values'] = data_V3['V1'].astype(str) + ';' + data_V3['V2'].astype(str) + ';' + data_V3['V3'].astype(str) + ';' + data_V3['V4'].astype(str) + ';' + data_V3['V5'].astype(str)          

data_V4 = data_V3.drop(['V1', 'V2', 'V3', 'V4', 'V5'], axis = 1)

data_V4
```
